```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

```python
school_data_complete=school_data_complete[['grade','school_name','reading_score','math_score','type','size','budget','Student ID']]
school_data_complete.head()
```

```python
school_names = school_data_complete['school_name'].unique()
school_names
```

```python
count_schools = len(school_names)
count_schools
```

```python
tot_students = school_data_complete['Student ID'].count()
tot_students
```

```python
schools = school_data_complete['school_name'].value_counts()
schools
```

```python
tot_budget = school_data_complete['budget'].unique().sum()
tot_budget
```

```python
avg_math = school_data_complete['math_score'].sum()/tot_students
avg_math
```

```python
avg_reading = school_data_complete['reading_score'].sum()/tot_students
avg_reading
```

```python
ps_math = school_data_complete['math_score'] >= 70
#Total number of students passing math
pass_math = np.sum(ps_math)
pass_math
```

```python
ps_read = school_data_complete['reading_score'] >= 70
#Total number of students passing reading
pass_reading = np.sum(ps_read)
pass_reading
```

```python
#Percent of students passing math
percent_math = pass_math/tot_students*100
percent_math
```

```python
#Percent of students passing reading
percent_read = pass_reading/tot_students*100
percent_read
```

```python
#Average percent of students passing math and reading
tot_rate = (percent_math + percent_read)/2
tot_rate
```

```python
dist_smry = {"Total Schools":[count_schools],
            "Total Students":[tot_students],
            "Total Budget":[tot_budget],
            "Average Math Score":[avg_math],
            "Average Reading Score":[avg_reading],
            "% Passing Math":[percent_math],
            "% Passing Reading":[percent_read],
            "% Overall Passing Rate":[tot_rate]
               }
dist_smry_df = pd.DataFrame(dist_smry)
dist_smry_df
```
